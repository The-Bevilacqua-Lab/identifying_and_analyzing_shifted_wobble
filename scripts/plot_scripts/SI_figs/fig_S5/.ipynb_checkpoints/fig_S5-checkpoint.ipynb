```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.spatial.distance import squareform
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster
import seaborn as sns
from collections import Counter
```

```python
ecoli_5S= pd.read_excel('../../../../results/reactivities_and_base_pairs_cleaned.xlsx', sheet_name='ecoli_5S_DMS')
ecoli_16S= pd.read_excel('../../../../results/reactivities_and_base_pairs_cleaned.xlsx', sheet_name='ecoli_16S_DMS')
ecoli_23S= pd.read_excel('../../../../results/reactivities_and_base_pairs_cleaned.xlsx', sheet_name='ecoli_23S_DMS')

yeast_5S= pd.read_excel('../../../../results/reactivities_and_base_pairs_cleaned.xlsx', sheet_name='yeast_5S_DMS')
yeast_58S= pd.read_excel('../../../../results/reactivities_and_base_pairs_cleaned.xlsx', sheet_name='yeast_58S_DMS')
yeast_18S= pd.read_excel('../../../../results/reactivities_and_base_pairs_cleaned.xlsx', sheet_name='yeast_18S_DMS')
yeast_25S= pd.read_excel('../../../../results/reactivities_and_base_pairs_cleaned.xlsx', sheet_name='yeast_25S_DMS')

hsapien_5S= pd.read_excel('../../../../results/reactivities_and_base_pairs_cleaned.xlsx', sheet_name='human_5S_DMS')
hsapien_18S= pd.read_excel('../../../../results/reactivities_and_base_pairs_cleaned.xlsx', sheet_name='human_18S_DMS')
hsapien_28S= pd.read_excel('../../../../results/reactivities_and_base_pairs_cleaned.xlsx', sheet_name='human_28S_DMS')
```

```python
dms_ecoli= pd.concat([ecoli_5S, ecoli_16S, ecoli_23S], ignore_index=True)
dms_ecoli.index= np.arange(0, len(dms_ecoli))

dms_yeast= pd.concat([yeast_5S, yeast_58S, yeast_18S, yeast_25S], ignore_index=True)
dms_yeast.index= np.arange(0, len(dms_yeast))

dms_hsapien= pd.concat([hsapien_5S, hsapien_18S, hsapien_28S], ignore_index=True)
dms_hsapien.index= np.arange(0, len(dms_hsapien))
```

```python
#ecoli
G_e= dms_ecoli[dms_ecoli['Sequence']=='G']
G_e.index= np.arange(0, len(G_e))

U_e= dms_ecoli[dms_ecoli['Sequence']=='U']
U_e.index= np.arange(0, len(U_e))

#yeast
G_y= dms_yeast[dms_yeast['Sequence']=='G']
G_y.index= np.arange(0, len(G_y))

U_y= dms_yeast[dms_yeast['Sequence']=='U']
U_y.index= np.arange(0, len(U_y))


#hsapien
G_h= dms_hsapien[dms_hsapien['Sequence']=='G']
G_h.index= np.arange(0, len(G_h))

U_h= dms_hsapien[dms_hsapien['Sequence']=='U']
U_h.index= np.arange(0, len(U_h))
```

```python
sns.set(style= 'ticks', font= "Arial")
fig, ax = plt.subplots(figsize = (8,8), dpi = 400)

G_e['norm_reactivity_nt'].hist(bins=50, color='#4daf4a', edgecolor='None', alpha= 0.5, density= True)
U_e['norm_reactivity_nt'].hist(bins=50, color='#a65628', edgecolor='None', alpha= 0.5, density= True)

#legend= plt.legend(loc='upper right', borderpad = 0, handletextpad = 0, prop={'size': 15},labels=['Reactivities of G','Reactivities of U'], frameon=False)

plt.grid(False)

plt.yticks(ticks=[0, 2, 4, 6, 8], fontsize = 15)
plt.ylim(-0.2, 9.5)

plt.tick_params(axis='x', labelsize=20)
plt.tick_params(axis='y', labelsize=20)

ax= plt.gca()
ax.spines.top.set_visible(False)
ax.spines.right.set_visible(False)

ax.set_xlabel("DMS reactivity of $\mathit{E. coli}$ rRNA", fontsize=20)
ax.set_ylabel("Density", fontsize=20)

#U677 reactivity
U677= U_e[(U_e['RNA']=='16S')&(U_e['Nucleotide']==677)]
U677_e= U677['norm_reactivity_nt'].to_list()[0]
plt.vlines(x= U677_e, ymin=0, ymax=7.5, color='#a65628', linewidth=1.5, linestyle='-')
plt.text(U677_e-0.02, 7.51, ' U677 (16S)', fontsize=20, rotation=90, color='#a65628')

#U1086 reactivity
U1086= U_e[(U_e['RNA']=='16S')&(U_e['Nucleotide']==1086)]
U1086_e= U1086['norm_reactivity_nt'].to_list()[0]
plt.vlines(x= U1086_e, ymin=0, ymax=7.5, color='#a65628', linewidth=1.5, linestyle='-')
plt.text(U1086_e-0.02, 7.58, 'U1086 (16S)', fontsize=20, rotation=90, color='#a65628')

#U2312 reactivity
U2312= U_e[(U_e['RNA']=='23S')&(U_e['Nucleotide']==2312)]
U2312_e= U2312['norm_reactivity_nt'].to_list()[0]
plt.vlines(x= U2312_e, ymin=0, ymax=7.5, color='#a65628', linewidth=1.5, linestyle='-')
plt.text(U2312_e-0.022, 7.58, 'U2312 (23S)', fontsize=15, rotation=90, color='#a65628')


#G713 reactivity
G713= G_e[(G_e['RNA']=='16S')&(G_e['Nucleotide']==713)]
G713_e= G713['norm_reactivity_nt'].to_list()[0]
plt.vlines(x= G713_e, ymin=0, ymax=7.5, color='#4daf4a', linewidth=1.5, linestyle='-')
plt.text(G713_e-0.02, 7.51, ' G713 (16S)', fontsize=20, rotation=90, color='#4daf4a')

#G1099 reactivity
G1099= G_e[(G_e['RNA']=='16S')&(G_e['Nucleotide']==1099)]
G1099_e= G1099['norm_reactivity_nt'].to_list()[0]
plt.vlines(x= G1099_e, ymin=0, ymax=7.5, color='#4daf4a', linewidth=1.5, linestyle='-')
plt.text(G1099_e-0.02, 7.58, 'G1099 (16S)', fontsize=20, rotation=90, color='#4daf4a')

#G2304 reactivity
G2304= G_e[(G_e['RNA']=='23S')&(G_e['Nucleotide']==2304)]
G2304_e= G2304['norm_reactivity_nt'].to_list()[0]
plt.vlines(x= G2304_e, ymin=0, ymax=7.5, color='#4daf4a', linewidth=1.5, linestyle='-')
plt.text(G2304_e-0.009, 7.55, 'G2304 (23S)', fontsize=15, rotation=90, color='#4daf4a')

plt.savefig('fig_S5A_Ecoli_rRNA_DMS_reactivity.pdf', format="pdf", bbox_inches="tight", dpi = 3000)
```
